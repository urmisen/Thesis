```python
#Binary_Encoding_Acetylation
from Bio import SeqIO
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile

def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  


def main():
    AA=[0]*31
    CC=[0]*31 
    DD=[0]*31
    EE=[0]*31 
    FF=[0]*31
    GG=[0]*31
    HH=[0]*31
    II=[0]*31
    KK=[0]*31
    LL=[0]*31 
    MM=[0]*31 
    NN=[0]*31
    PP=[0]*31
    QQ=[0]*31 
    RR=[0]*31 
    SS=[0]*31 
    TT=[0]*31 
    VV=[0]*31 
    WW=[0]*31 
    YY=[0]*31 
    XX=[0]*31
    K_index=[]
    K_index_N=[]
    for_value=[]
    all_value=[]
    truncated=[]
    full_truncated=[]
    Seq=[]
    lst=[]
    Acet=[]
    yes_index=[]
    New_yes=[]
    
    
    #fasta sequence read with SeqIQ.parse
    records = list(SeqIO.parse("Acetylation_Homo.fasta.txt", "fasta"))
#     print(records)
#putting the fasta sequences(seq_record.seq) in sequence list(sequence=[])
    sequence=[]
    for seq_record in SeqIO.parse("Acetylation_Homo.fasta.txt", "fasta"):
        sequence.append(seq_record.seq)

    # print (len(sequence))
#     print(sequence)
    length=len(sequence)
#putting the fasta protein names(seq_record.name) in names list(names=[])
    names=[]
    for seq_record in SeqIO.parse("Acetylation_Homo.fasta.txt", "fasta"):
        name=seq_record.name.split("|")[1]
        names.append(name)
    # print(names)
    names_len=len(names)
    
    
# Window creation of size 31

    for seq_record in range(0,length):
        S=list(sequence[seq_record])
        Seq.append(S)
#         print(Seq)
#finding the occurrences of K and its indexes in every sequence
        occurrences = lambda s, lst: (i for i,e in enumerate(lst) if e == s)
        K = list(occurrences('K', S))
        K_new=list(map(lambda x: x + 1, K))
        K_index_N.append(K_new)
        K_index.append(K)
#K_index contains the the list like [[2,34,56,78],[3,6,9]]    
#K_value contains every list of list like [2,34,56,78] for [[2,34,56,78],[3,6,9]] 
#S_value contains full on fasta sequence
        
    K_lenth=len(K_index)                   
    for i in range(0,K_lenth):
        K_value=K_index[i]
        S_value=Seq[i]
        max_len=len(S_value)
#         print(names[i])
#         print(K_value)
        for j in K_value:
            
            if (j-15>0) and (j+15<max_len):
                X=S_value[j-15:j+16]
            elif (j-15<0) and (j+15<max_len):
                Neg=15-j
                
                X1=['X']*Neg
                X2=S_value[0:j]
                X3=list(S_value[j])
                X4=S_value[j+1:j+16]
                X=X1+X2+X3+X4
            elif (j-15>0) and (j+15>max_len):
                pos=(j+15)-(max_len-1)
                X1=S_value[j-15:j]
                X2=list(S_value[j])
                X3=S_value[j+1:max_len]
                X4=['X']*pos
                X=X1+X2+X3+X4
#truncated list contains the 31long peptide lists in a list like [['X','X','X','R','A'],['A','C','V,'W' ,'T']] 
#full_truncated contains the 31long peptide lists in a list as string like [[XXXRA],[ACVWT]]
            
#             Str= listToString(X)
            truncated.append(X)
            
#             Str= listToString(Str_X)            
#             f.write(Str)
#         f.write("\n")
#     print(truncated)
    truncated_len=len(truncated)
    
    for i in range(0,truncated_len):
#         unique=list(set(truncated[i]))
#         for j in unique:

#             occurrences = lambda s, lst: (k for k,e in enumerate(lst) if e == s)
#             N = list(occurrences(j, truncated[i]))
            
        df = pd.DataFrame(
            {'Peptide': truncated[i]
             }
        )
        df['A'] = df['Peptide'].apply(lambda x: 1 if x == 'A' else 0)
        df['C'] = df['Peptide'].apply(lambda x: 1 if x == 'C' else 0)
        df['D'] = df['Peptide'].apply(lambda x: 1 if x == 'D' else 0)
        df['E'] = df['Peptide'].apply(lambda x: 1 if x == 'E' else 0)
        df['F'] = df['Peptide'].apply(lambda x: 1 if x == 'F' else 0)
        df['G'] = df['Peptide'].apply(lambda x: 1 if x == 'G' else 0)
        df['H'] = df['Peptide'].apply(lambda x: 1 if x == 'H' else 0)
        df['I'] = df['Peptide'].apply(lambda x: 1 if x == 'I' else 0)
        df['K'] = df['Peptide'].apply(lambda x: 1 if x == 'K' else 0)
        df['L'] = df['Peptide'].apply(lambda x: 1 if x == 'L' else 0)
        df['M'] = df['Peptide'].apply(lambda x: 1 if x == 'M' else 0)
        df['N'] = df['Peptide'].apply(lambda x: 1 if x == 'N' else 0)
        df['P'] = df['Peptide'].apply(lambda x: 1 if x == 'P' else 0)
        df['Q'] = df['Peptide'].apply(lambda x: 1 if x == 'Q' else 0)
        df['R'] = df['Peptide'].apply(lambda x: 1 if x == 'R' else 0)
        df['S'] = df['Peptide'].apply(lambda x: 1 if x == 'S' else 0)
        df['T'] = df['Peptide'].apply(lambda x: 1 if x == 'T' else 0)
        df['V'] = df['Peptide'].apply(lambda x: 1 if x == 'V' else 0)
        df['W'] = df['Peptide'].apply(lambda x: 1 if x == 'W' else 0)
        df['Y'] = df['Peptide'].apply(lambda x: 1 if x == 'Y' else 0)
        df['X'] = df['Peptide'].apply(lambda x: 1 if x == 'X' else 0)
            
#             df.to_csv('/Users/urmis/Documents/Thesis/Binary_Encoding.csv',index=False)
            

        print(df)
                
        
                      

    
    

main()
```
